```python
## 离散化处理（分箱） 
```

```python
import pandas as pd
ages = [20,21,23,25,28,30,37,62,69,55,35,120]
# 幼年 青年 壮年 老年
bins = [0,15,30,40,121]

# ages 代表你要分箱的数据
# bins 给定区间边界值
# labels 给分箱命名
# right 选择边界包含方向
print(pd.cut(ages,bins=bins,right=False).value_counts())
pd.cut(ages,bins=bins,right=False,labels=['幼年','青年','壮年','老年']).value_counts()
```

```python
import numpy as np
# 等距分箱
data = np.random.randint(1,30,30)
# 4 代表分成几个区间
print(pd.cut(data,4).value_counts())

# 等频分箱
data = np.random.randint(1,100,1000)
pd.qcut(data,5).value_counts()
```

## Matplotlib 

```python
from matplotlib import pyplot as plt
months = ['January','February','March','April','May','June','July','August','September','October','November','December']
sales = [300,240,270,310,300,350,310,316,323,307,420,370]
```

## 折线图

```python
# 指写画布大小
plt.figure(figsize=(12,4))
# 引入折线图的方法
plt.plot(months,sales)
# 显示 pycharm 中 必须要写显示方法
plt.show()
```

```python
月份 = ['January','February','March','April','May','June','July','August','September','October','November','December']
销售额 = [300,240,270,310,300,350,310,316,323,307,420,370]

# 设置字体
plt.rcParams['font.family']='SimHei' # windows系统-- SimHei??
plt.rcParams['font.size']=11
# 指写画布大小
plt.figure(figsize=(12,4),dpi=100,facecolor=(0.5,1,0)) # 'red'
# <dots per inch> 每英寸的像素点数
# 引入折线图的方法
plt.plot(
    月份, # x轴数据
    销售额, # y轴数据
    marker='o', # 标点的形状 'D' 方点
    markerfacecolor='yellow', # 标点的颜色
    markeredgecolor='red', # 标点边框的颜色
    markersize='10', # 标点大小
    linestyle='-.', # 线段样式
    color='red', # 线段的颜色
    linewidth=2, # 也可以用lw=2，线段的粗细
    alpha=0.5 # 线段的透明度 0-1 之间
)
# x轴的标签
plt.xlabel('月份')
# y轴的标签
plt.ylabel('销售/万')
# 标题
plt.title('全年销售趋势折线图')
# 网格
plt.grid(color='gray',linestyle=':',alpha=0.5,linewidth=1)
# 保存图片
plt.savefig('全年销售趋势-折线图.png')
# 显示 pycharm 中 必须要写显示方法
plt.show()
```

```python
# 指写画布大小
plt.figure(figsize=(12,4),dpi=100,facecolor=(0.5,1,0))
plt.scatter(
    月份,
    销售额,
    # width=0.5,
)
plt.show()
```

```python
年份 = [2015,2016,2017,2018,2019,2020,2021,2022,2023]
逾期率 = [0.0136,0.0135,0.0125,0.0118,0.0147,0.0108,0.0101,0.0098,0.0123]
金额 = [388.8,406.6,459.7,788.6,837.9,838.6,858.6,917.1,990.04]
```

```python
# 创建新的图表
fig,axl = plt.subplots(figsize=(12,4))
print(fig,axl)

# 绘制逾期率的折线图
# axl.plot(年份,逾期率)
axl.plot(年份,逾期率,'b-o',label='逾期率')
"""
b-o
b 代表的是颜色
- 代表的是线条的样式
o 代表的是标点的形状
"""
axl.set_xlabel('年份')
axl.set_ylabel('逾期率',color='b')

# 创建右边的y轴
raxl = axl.twinx()
# 绘制金额的折线图
raxl.plot(年份,金额,'g:^',label='金额')
raxl.set_ylabel('逾期金额/亿元',color='g')
# 为每一个标点添加上数据标
for i in range(len(年份)):
    axl.text(年份[i],逾期率[i],f'{逾期率[i]*100:.2f}%',color='b',ha='left')
    raxl.text(年份[i],金额[i],f'{金额[i]:.2f}亿',color='g',ha='right')
# 标题
plt.title('2015-2023信用卡逾期数据')
# 网络
plt.grid(color='gray',linestyle=':',alpha=0.5,linewidth=1)
# 添加图例
lines1, aaa = axl.get_legend_handles_labels()
lines2, bbb = raxl.get_legend_handles_labels()
axl.legend(lines1+lines2,aaa+bbb,loc='upper left')

plt.show
```
